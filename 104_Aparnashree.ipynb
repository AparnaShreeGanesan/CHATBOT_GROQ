```python
#getting input from the user
n=int(input("enter a number "))
#initialize fib0=0,fib1=1
fib0=0
fib1=1
#initialize i=1
# 'i' keeps track of how many terms have been printed in the series
i=1
print("Fibnocci series is")
#loop to generate fibnocci series of n terms
while i<=n:
  #print current fibnocci number with space
  print(fib0, end=' ')
  fib0,fib1=fib1,fib0+fib1
  #increment i value
  i=i+1
```

```python
# Find the largest element in a list using boolean comparison

lst = [10, 25, 3, 89, 45]

largest = lst[0]

for i in lst:
    is_greater = i > largest   # Boolean comparison expression
    if is_greater:
           largest = i

print("Largest element:", largest)
```
